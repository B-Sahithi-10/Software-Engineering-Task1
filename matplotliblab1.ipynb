```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5, 6, 7]
y = [5, 15, 20, 25, 35, 45, 55]

plt.plot(x, y, marker='o', linestyle='--', color='r', label='Data Trend')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Enhanced Line Plot with Customizations')

plt.grid(True)

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```
